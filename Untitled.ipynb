```python
import rlcard
import collections
import random
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
learning_rate = 0.0005
gamma = 0.98
buffer_limit = 50000
batch_size = 32
env = rlcard.make('no-limit-holdem')
s=env.reset()
cnt=0
print(s)
print('---------------------------------------------------------------------------')
while(cnt<20):
    s=env.step(1)
    print(s)
    print('---------------------------------------------------------------------------')

    cnt+=1




print(s)
#print(env.state_shape)
#print(env.actions.FOLD.value)
```

```python
class ReplayBuffer():
    def __init__(self):
        self.buffer = collections.deque(maxlen=buffer_limit)    # double-ended queue
    
    def put(self, transition):
        self.buffer.append(transition)

    def sample(self, n):
        mini_batch = random.sample(self.buffer, n)
        s_lst, a_lst, r_lst, s_prime_lst, done_mask_lst = [], [], [], [], []

        for transition in mini_batch:
            s, a, r, s_prime, done_mask = transition
            s_lst.append(s)
            a_lst.append([a])
            r_lst.append([r])
            s_prime_lst.append(s_prime)
            done_mask_lst.append([done_mask])

        return torch.tensor(s_lst, dtype=torch.float), torch.tensor(a_lst), \
               torch.tensor(r_lst), torch.tensor(s_prime_lst, dtype=torch.float), \
               torch.tensor(done_mask_lst)

    def size(self):
        return len(self.buffer)
```

```python
class Qnet(nn.Module):
    def __init__(self):
        super(Qnet, self).__init__()
        self.fc1 = nn.Linear(54, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 5)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
      
    def sample_action(self, state, epsilon):

        print('#@!?@!#?@!?#@!#?@!')
        print(state)
        if type(state) is tuple:
            obs=torch.from_numpy(state[0]['obs']).float()
        else:
            obs=torch.from_numpy(state['obs']).float()
        out = self.forward(obs)
        coin = random.random()
        if coin < epsilon:
            a=random.randint(0,5)
        else : 
            return out.argmax().item()   
```

```python
def train(q, q_target, memory, optimizer):
    for i in range(10):
        s,a,r,s_prime,done_mask = memory.sample(batch_size)

        q_out = q(s)
        q_a = q_out.gather(1,a)

        # DQN
        max_q_prime = q_target(s_prime).max(1)[0].unsqueeze(1)


        target = r + gamma * max_q_prime * done_mask
        
        # MSE Loss
        loss = F.mse_loss(q_a, target)

        # Smooth L1 Loss
        #loss = F.smooth_l1_loss(q_a, target)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
def main():
    env = rlcard.make('no-limit-holdem')
    q = Qnet()
    q_target = Qnet()
    q_target.load_state_dict(q.state_dict())
    memory = ReplayBuffer()

    print_interval = 20
    score = 0.0
    optimizer = optim.Adam(q.parameters(), lr=learning_rate)
    cnt=0

    for n_epi in range(3000):
        epsilon = max(0.01, 0.08 - 0.01*(n_epi/200)) #Linear annealing from 8% to 1%
        s = env.reset()
        done = False
        while not done:
            a = q.sample_action(s, epsilon)

            print(a)
            s_prime, done = env.step(a)
            done_mask = 0.0 if done else 1.0
            memory.put((s,a,r/100.0,s_prime, done_mask))
            s = s_prime

            score += r
            cnt+=1
            if done:
                break

        if memory.size()>2000:
            train(q, q_target, memory, optimizer)

        if n_epi%print_interval==0 and n_epi!=0:
            q_target.load_state_dict(q.state_dict())
            print("n_episode :{}, score : {:.1f}, n_buffer : {}, eps : {:.1f}%".format(
                                                            n_epi, score/print_interval, memory.size(), epsilon*100))
            score = 0.0

    env.close()
```

```python
if __name__ == '__main__':
    main()
```
